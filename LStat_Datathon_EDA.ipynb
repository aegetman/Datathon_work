```python
import pandas as pd
import pyarrow
import os
```

```python
# Get the current directory
current_dir = os.getcwd()

# List all files in the directory
files = os.listdir(current_dir)

# Filter out the Parquet files
parquet_files = [file for file in files if file.endswith('.parquet.gzip') or file.endswith('.parquet')]

# Loop through each Parquet file, read it, and assign to a DataFrame
for file in parquet_files:
    # Construct the full file path
    file_path = os.path.join(current_dir, file)
    
    # Extract dataframe name from file name without extension
    dataframe_name = os.path.splitext(file)[0]
    # Remove the extension .gzip if present
    dataframe_name = dataframe_name.replace('.gzip', '')
    # Remove the extension .parquet if present
    dataframe_name = dataframe_name.replace('.parquet', '')
    # Add the _df suffix
    dataframe_name += '_df'
    
    # Read the Parquet file into a DataFrame and assign it to a dynamically created variable
    globals()[dataframe_name] = pd.read_parquet(file_path)

    # Now you have individual variables named interventions_df, x_df, etc., each containing a DataFrame
    # You can directly access these DataFrames and perform operations on them, for example:
    print(f"DataFrame '{dataframe_name}' shape: {globals()[dataframe_name].shape}")
    print(globals()[dataframe_name].head())  # Example operation on the DataFrame
```

```python
from datetime import datetime

#format_str_1 =
format_str = "%d%b%y:%H:%M:%S"  # Pandas can not automatically convert this

def parse_datetime_columns(df, columns, format_str=None):
    for col in columns:
        if format_str:
            df[col] = pd.to_datetime(df[col], format=format_str, errors='coerce')
        else:
            df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
columns_to_parse = ['t1', 't2', 't3', 't4', 't5','t6','t7','t9', 't0']
parse_datetime_columns(cardiac_interventions_bxl_df, columns_to_parse)        
cardiac_interventions_bxl_df
```

```python
# Calculate average speed and time_difference_seconds
cardiac_interventions_bxl_df['time_difference_seconds'] = cardiac_interventions_bxl_df['time_difference'].dt.total_seconds()
cardiac_interventions_bxl_df['average_speed'] = cardiac_interventions_bxl_df['calculated_distance_departure_to'] / cardiac_interventions_bxl_df['time_difference_seconds']
```

```python
# Group by streetname and calculate count of late interventions, average speed, and average time difference
latency_threshold= 15

grouped_df = cardiac_interventions_bxl_df.groupby('streetname_permanence').agg(
    late_interventions_count=('time_difference', lambda x: (x > pd.Timedelta(minutes=latency_threshold)).sum()),
    average_speed=('average_speed', 'mean'),
    average_time_difference=('time_difference_seconds', 'mean')
).reset_index()
```

```python
grouped_df
```

```python
# Define the latency threshold
latency_threshold = 15

# Calculate initial count of interventions for each street
initial_count = cardiac_interventions_bxl_df.groupby('streetname_permanence').size().reset_index(name='initial_count')

# Calculate count of late interventions for each street
late_interventions_count = cardiac_interventions_bxl_df[cardiac_interventions_bxl_df['time_difference'] > pd.Timedelta(minutes=latency_threshold)]
late_interventions_count = late_interventions_count.groupby('streetname_permanence').size().reset_index(name='late_interventions_count')

# Calculate average speed for each street
average_speed_by_street = cardiac_interventions_bxl_df.groupby('streetname_permanence')['average_speed'].mean().reset_index()

# Calculate average time difference for each street
average_time_difference_by_street = cardiac_interventions_bxl_df.groupby('streetname_permanence')['time_difference'].mean().reset_index(name='average_time_difference')

# Merge initial count and late interventions count
merged_df = pd.merge(initial_count, late_interventions_count, on='streetname_permanence', how='left')

# Merge average speed and average time difference
merged_df = pd.merge(merged_df, average_speed_by_street, on='streetname_permanence', how='left')
merged_df = pd.merge(merged_df, average_time_difference_by_street, on='streetname_permanence', how='left')

# Fill missing values with 0 (no late interventions)
merged_df['late_interventions_count'].fillna(0, inplace=True)

# Calculate proportion of late interventions to total interventions
merged_df['proportion_late_interventions'] = merged_df['late_interventions_count'] / merged_df['initial_count']

merged_df
```
